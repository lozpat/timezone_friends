```python
import datetime
import pytz #timezones module
import pandas as pd
import numpy as np

my_tz = 'Europe/London' #INPUT TIMEZONE, IN PYTZ FORMAT - CHOOSE FROM DROPDOWN
their_tz = 'Australia/Sydney' #INPUT TIMEZONE, IN PYTZ FORMAT - CHOOSE FROM DROPDOWN
my_early_time_limit = datetime.time(10, 0, 0) #INPUT EARLIEST AVAILABLE TIME
my_late_time_limit = datetime.time(22, 0, 0) #INPUT LATEST AVAILABLE TIME
their_early_time_limit = datetime.time(18, 0, 0) #INPUT EARLIEST AVAILABLE TIME
their_late_time_limit = datetime.time(22, 0, 0) #INPUT LATEST AVAILABLE TIME

if my_tz in pytz.all_timezones:
    my_dt = datetime.datetime.now(pytz.timezone(my_tz))
    print('My time:', my_dt)
    my_offset = my_dt.utcoffset()
else: print('Incorrect timezone, check entry you willy')

if their_tz in pytz.all_timezones:
    their_dt = datetime.datetime.now(pytz.timezone(their_tz))
    print('Their time:', their_dt)
    their_offset = their_dt.utcoffset()
else: print('Incorrect timezone, check entry you willy')

if my_offset > their_offset: #need to account for same time zone
    time_difference = my_offset - their_offset
    preposition = 'behind'
else: time_difference = their_offset - my_offset
preposition = 'ahead'

print('They are', time_difference, preposition) #assess time offset behind/ahead issue
```

```python
my_time_midnight = (datetime.datetime.now().replace(hour=0, minute=0, second=0, microsecond=0))
my_time = [my_time_midnight.time()]
their_time_midnight = my_time_midnight + time_difference
their_time = [their_time_midnight.time()]
timestep = datetime.timedelta(hours=1)

for i in range (1, 24):
    my_time.append((my_time_midnight + i * timestep).time())

for i in range (1, 24):
    their_time.append((their_time_midnight + i * timestep).time())

combined_times = {'My time': my_time,
                  'Their time': their_time
                 }

df = pd.DataFrame(combined_times)
df
```

```python
my_early_index_limit = df.loc[df['My time'] == my_early_time_limit].index
my_late_index_limit = df.loc[df['My time'] == my_late_time_limit].index

df_truncate = df.iloc[my_early_index_limit[0]:my_late_index_limit[0]+1].reset_index(drop=True)
df_truncate
```

```python
their_early_index_limit = df_truncate.loc[df_truncate['Their time'] == their_early_time_limit].index
their_late_index_limit = df_truncate.loc[df_truncate['Their time'] == their_late_time_limit].index

df_availability = df_truncate.iloc[their_early_index_limit[0]:their_late_index_limit[0]+1].reset_index(drop=True)
df_availability
```
